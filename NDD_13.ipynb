```python
import pandas as pd

df1 = pd.read_excel('xlsx_/f_029df (1).xlsx')
df2 = pd.read_excel('xlsx_/f_9218df (1).xlsx')
df3 = pd.read_excel('xlsx_/f_18224df (1).xlsx')
```

```python
df1.head()
```

```python
df1 = df1.drop('Unnamed: 0', axis = 1)
df1 = df1.drop('Unnamed: 0.1', axis = 1)
df1 = df1.drop('X', axis = 1)
df1 = df1.drop('Y', axis = 1)

df2 = df2.drop('Unnamed: 0', axis = 1)
df2 = df2.drop('Unnamed: 0.1', axis = 1)
df2 = df2.drop('X', axis = 1)
df2 = df2.drop('Y', axis = 1)

df3 = df3.drop('Unnamed: 0', axis = 1)
df3 = df3.drop('Unnamed: 0.1', axis = 1)
df3 = df3.drop('X', axis = 1)
df3 = df3.drop('Y', axis = 1)
```

```python
# 0 to 9시 데이터와 18시 to 24시 데이터를 합치고 평균내서 새로운 데이터프레임 생성

df = pd.concat([df1, df3], ignore_index=True).groupby('dong').mean()
df = df.reset_index()
df
```

```python
# 총생활인구수를 마지막 칼럼에 위치하게끔하겠습니다.
# 먼저 a로 총생활인구수 값들을 받아줍니다.
a = df['총생활인구수'].values
type(a)
```

```python
# 총생활인구수 칼럼을 drop 시키고 population이라는 column을 추가합니다.
df = df.drop('총생활인구수', axis = 1)
df['population'] = a
```

```python
df.head()
```

```python
# 전에 heatmap을 그린거 보면 ffc는 데이터가 너무 없어서 -값 내지는 매우 작은값을 보였다.
# 머신러닝돌릴때 ffc와 affc도 drop하자.
df = df.drop('ffc', axis = 1)
df = df.drop('affc', axis = 1)
```

```python
# null값을 0으로 채워둡니다.
df = df.fillna(0)
df = df.drop(columns=['nfnc','nfc','nsnc','nsc','anfnc','anfc','ansnc','ansc'])
```

```python
df.to_excel('xlsx_/nonfullDF.xlsx', index=False)
```

```python
a = df2['총생활인구수'].values
df2 = df2.drop('총생활인구수', axis = 1)
df2['population'] = a
df2 = df2.drop('ffc', axis = 1)
df2 = df2.drop('affc', axis = 1)
df2 = df2.fillna(0)
df2.head()
```

```python
df2.to_excel('xlsx_/fulltimeDF.xlsx', index=False)
```

```python
# 다중선형회귀 돌리기
# 먼저 데이터 셋을 트레인 셋과 test set으로 분리시키자.
!pip install tensorflow
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf

seed = 100
np.random.seed(seed)
tf.random.set_seed(seed)
data = df.values
x = data[:,:-1].astype(float)
y = data[:,-1].astype(float)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state = seed)
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(7,input_dim = 7, activation='sigmoid'))
model.add(Dense(3,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.fit(x_train, y_train, epochs = 100, batch_size = 10)
```

```python
y_predict = model.predict(x_test)

import matplotlib.pyplot as plt
plt.scatter(y_test, y_predict, alpha=0.4)
plt.plot([0,1],[0,1], c = 'red')
plt.xlabel("Actual population")
plt.ylabel("Predicted population")
plt.title("Deep Neural Network dense")
plt.show()
# plt.savefig('DNN_test.jpg')
```

```python
# plot all data
y_predict = model.predict(x)

import matplotlib.pyplot as plt
plt.scatter(y, y_predict, alpha=0.4)
plt.plot([0,1],[0,1], c = 'red')
plt.xlabel("Actual population")
plt.ylabel("Predicted population")
plt.title("Deep Neural Network dense")
plt.show()
# plt.savefig('DNN_full.jpg')
```
